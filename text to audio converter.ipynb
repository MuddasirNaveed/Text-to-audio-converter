```python
from gtts import gTTS
import os

def text_to_speech(input_file, output_audio_file, language='en'):
    """
    Converts a text file into speech and saves it as an audio file.
    
    Parameters:
        input_file (str): Path to the input text file containing captions.
        output_audio_file (str): Path to the output audio file.
        language (str): Language code for the text-to-speech (default is 'en' for English).
    """
    try:
        # Read the text from the input file
        with open(input_file, 'r', encoding='utf-8') as file:
            text = file.read().strip()
        
        if not text:
            print("The text file is empty.")
            return

        # Convert text to speech
        tts = gTTS(text=text, lang=language, slow=False)

        # Save the audio file
        tts.save(output_audio_file)
        print(f"Audio file saved as: {output_audio_file}")
    
    except Exception as e:
        print(f"An error occurred: {e}")

# Example Usage
input_text_file =r"C:\Users\muddasir.naveed\Downloads\1.txt"
output_audio_file =r"output_audio.mp3"  # Replace with the desired output file name

text_to_speech(input_text_file, output_audio_file)
```

```python
import os
import pyttsx3

def text_to_speech(input_folder, output_folder):
    try:
        # Initialize the pyttsx3 engine
        engine = pyttsx3.init()

        # Get the list of voices and set the male voice
        voices = engine.getProperty('voices')
        engine.setProperty('voice', voices[0].id)  # Index 0 is typically a male voice

        # Set the rate (speed) of the speech (optional)
        engine.setProperty('rate', 150)

        # Ensure the output folder exists
        if not os.path.exists(output_folder):
            os.makedirs(output_folder)

        # Iterate over all the .txt files in the input folder
        for filename in os.listdir(input_folder):
            if filename.endswith('.txt'):
                input_file = os.path.join(input_folder, filename)
                output_audio_file = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}.mp3")

                # Read the text from the input file
                with open(input_file, 'r', encoding='utf-8') as file:
                    text = file.read().strip()

                if not text:
                    print(f"The file {filename} is empty.")
                    continue

                # Convert text to speech and save as an MP3 file
                engine.save_to_file(text, output_audio_file)
                engine.runAndWait()
                print(f"Audio file saved as: {output_audio_file}")

    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
input_folder = r"C:\Users\muddasir.naveed\Downloads\`voice over for slides"  # Change to your input folder path
output_folder = r"C:\Users\muddasir.naveed\Downloads\voice-over for slides"  # Change to your output folder path

text_to_speech(input_folder, output_folder)
```

```python
from google.cloud import texttospeech

def text_file_to_speech(input_file, output_audio):
    # Initialize the Text-to-Speech client
    client = texttospeech.TextToSpeechClient()

    # Read the input text from the file
    with open(input_file, "r") as file:
        text = file.read()

    # Set the text input to be synthesized
    synthesis_input = texttospeech.SynthesisInput(text=text)

    # Configure the voice settings (e.g., language code, voice name, gender)
    voice = texttospeech.VoiceSelectionParams(
        language_code="en-US",
        name="en-US-Neural2-J",  # Adjust for specific voice
        ssml_gender=texttospeech.SsmlVoiceGender.MALE,
    )

    # Configure the audio output settings
    audio_config = texttospeech.AudioConfig(
        audio_encoding=texttospeech.AudioEncoding.MP3
    )

    # Perform the Text-to-Speech request
    response = client.synthesize_speech(
        input=synthesis_input, voice=voice, audio_config=audio_config
    )

    # Save the audio to the output file
    with open(output_audio, "wb") as out:
        out.write(response.audio_content)
    print(f"Audio content written to {output_audio}")
#input_text_file =r"C:\Users\muddasir.naveed\Downloads\slide2.1.txt"

# Usage
input_file = r"C:\Users\muddasir.naveed\Downloads\slide2.1.txt"  # Your input text file
output_audio = r"C:\Users\muddasir.naveed\Downloads\slide2.1.mp3"  # Your desired output audio file

#output_audio_file = r"C:\Users\muddasir.naveed\Downloads\slide2.1.mp3"

text_file_to_speech(input_file, output_audio)
```

```python
from TTS.api import TTS

def text_to_speech(input_file, output_file, model_name="tts_models/en/vctk/vits"):
    """
    Convert text from a file to human-like speech (male voice) and save it as an audio file.

    Args:
        input_file (str): Path to the text file.
        output_file (str): Path to save the generated audio file.
        model_name (str): Coqui TTS model for male voice.
    """
    try:
        # Load the pre-trained TTS model (male voice)
        print("Loading TTS model...")
        tts = TTS(model_name)
        
        # Read the text file
        with open(input_file, "r", encoding="utf-8") as file:
            text = file.read()
        
        # Generate speech and save to a file
        print("Generating speech...")
        tts.tts_to_file(text=text, file_path=output_file)
        
        print(f"Speech saved to: {output_file}")
    
    except Exception as e:
        print(f"Error: {e}")

# Example Usage
if __name__ == "__main__":
    input_file = r"C:\Users\muddasir.naveed\Downloads\slide2.1.txt"  # Your input text file
    output_audio = r"C:\Users\muddasir.naveed\Downloads\slide2.1.mp3"  # Your desired output audio file
    text_to_speech(input_text_file, output_audio_file)
```
